```python
df = pd.read_csv("./data/XLE_energy_etf.csv")
df.head() # Ensure stats are successfully read
```

```python
df.describe() # Shows basic statistics for data
```

```python
df['Date'] = pd.to_datetime(df['Date'], infer_datetime_format=True)
indexeddf = df.set_index(['Date']) # Passed strings into a datetime format using to_datetime (inbuilt function in pandas)
```

```python
from datetime import datetime
indexeddf.head(5) # Dates should be bold now and the word \"Date\" should be of a different row from the rest
```

```python
# Test plot using \"High\" of stock
dfhigh = indexeddf['High']
plt.xlabel("Date")
plt.ylabel("High of Stock")
plt.plot(dfhigh)
```

```python
df10to12 = dfhigh.loc['2010-01-04':'2012-01-01'] # filtering out data up till 2012 to have a better idea of the plot
df10to12.describe()
plt.xlabel("Date")
plt.ylabel("High of Stock")
plt.plot(df10to12)
```

```python
rolmean = dfhigh.rolling(window=365).mean() # Sets up the rolling window forecasts (in this case set to 1 year)
rolstd = dfhigh.rolling(window=365).std() # Forecast for residual SEs
rolema = indexeddf.High.ewm(span=365, adjust=False).mean() # Uses EMA (exponential moving average)
# Both uses standard MA formulas

print(rolmean, rolstd) # You would see that the first year is NaN because no prediction avaliable for that
```

```python
original = plt.plot(dfhigh, color='blue', label='Original')
mean = plt.plot(rolmean, color='red', label='Rolling Mean')
resi = plt.plot(rolstd, color='black', label='Rolling Residuals')
ema = plt.plot(rolema, color='green', label='Rolling EMA')
plt.legend(loc='best')
plt.title('Rolling Mean & Residuals')
plt.show(block=False)
```
